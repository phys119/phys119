This catch-up lab uses a simulation that duplicates an in-person experiment where you place a large spring on a mass balance and use your hand to compress the spring, using a ruler to measure the length of the spring and the mass balance to measure the force applied to the spring.

Hooke's Law tells us that the spring constant $k$ is the proportionality constant of a particular spring that relates the change in force ($\Delta F$) needed to extend or compress a spring by some distance ($\Delta x$):

$$k = \frac{|\Delta F|}{|\Delta x|} = g\frac{|m_2 - m_1|}{|x_2 - x_1|}$$

where:

- $x_1$ is the position of the top of the spring when it is uncompressed,
- $x_2$ is the position of the top of the spring when it is compressed,
- $m_1$ is the mass displayed on the scale when the spring is uncompressed, and
- $m_2$ is the mass displayed on the scale when the spring is compressed.

The mass $m$ displayed on the scale can be related to force $F$ by $F = mg$, where $g$ is the acceleration due to gravity at our altitude in Vancouver: $g = 9.809$ m/s².

## Appendix: Uncertainty Propagation Equations

The following equations help you go from single-measurement uncertainties $\delta m_1$, $\delta m_2$, $\delta x_1$, $\delta x_2$ to an uncertainty $\delta k$ in your spring constant. Many paths lead to $\delta k$; no single path requires all of these equations.

### *Propagation Formulas*

| Formula | Uncertainty Propagation |
|---------|------------------------|
| $k = \frac{|\Delta F|}{|\Delta x|}$ | $\delta k = k\sqrt{\left(\frac{\delta(\Delta F)}{\Delta F}\right)^2 + \left(\frac{\delta(\Delta x)}{\Delta x}\right)^2}$ |
| $|\Delta x| = |x_2 - x_1|$ | $\delta(\Delta x) = \sqrt{(\delta x_1)^2 + (\delta x_2)^2}$ |
| $|\Delta F| = |F_2 - F_1|$ | $\delta(\Delta F) = \sqrt{(\delta F_1)^2 + (\delta F_2)^2}$ |
| $|\Delta F| = g|\Delta m|$ | $\delta(\Delta F) = g \cdot \delta(\Delta m)$ |
| $F = g \cdot m$ | $\delta F = g \cdot \delta m$ |
| $|\Delta m| = |m_2 - m_1|$ | $\delta(\Delta m) = \sqrt{(\delta m_1)^2 + (\delta m_2)^2}$ |
