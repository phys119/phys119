### *Relative Uncertainty Review*

(The following expands on the slides from this section)

Given a measurement $\pm \delta A$ the formula for relative uncertainty is:

> Relative uncertainty = $\delta A / A$

Notes on relative uncertainty

* Absolute uncertainty (δA, as opposed to δA/A) provides minimal information about the precision of the measurement. When we use only absolute uncertainty, it can be hard to compare the precision of two measurements. 
* For example: Both the measurements m1 = (17 ± 10) kg and m2 = (103 ± 10) kg have the same absolute uncertainties (10 kg), but the relative uncertainties are quite different
    - m1 relative uncertainty = 0.59 (59%)
    - m2 relative uncertainty = 0.09 (9%)
* Relative uncertainty and precision refer to the same concept, but with opposite directionality from each other. A higher precision / more precise measurement refers to one with a lower relative uncertainty.
* Throughout this course, we are going to try to improve our measurement strategies to improve the quality of our data, primarily by getting more precise measurements. To determine if our measurement strategy improvement worked as we expected, we use relative uncertainty to compare, as the actual value of our measurement might change each time.

## Appendix 1 - Relative Uncertainty (also known as Precision)

Consider a measurement reported as $A \pm \delta A$.

The **relative uncertainty** is the ratio $\delta A / A$.

Remember:

- Uncertainty on its own can&rsquo;t be &ldquo;good&rdquo; or &ldquo;bad&rdquo; without knowing how big it is relative to the measurement;
- A more precise measurement will have a smaller relative uncertainty.

## Appendix 2 - Uncertainty Propagation Equations and LaTeX Code

The following equations help you go from your single-measurement uncertainties $\delta m_1$, $\delta m_2$, $\delta x_1$ and $\delta x_2$ to an uncertainty $\delta k$ in your spring constant. Many paths lead to $\delta k$; no single path requires all of these equations.

### *Propagation Formulas*

| Formula | Uncertainty Propagation |
|---|---|
| $k = \frac{|\Delta F|}{|\Delta x|}$ | $\delta k = k\sqrt{\left(\frac{\delta(\Delta F)}{\Delta F}\right)^2 + \left(\frac{\delta(\Delta x)}{\Delta x}\right)^2}$ |
| $|\Delta x| = |x_2 - x_1|$ | $\delta(\Delta x) = \sqrt{(\delta x_1)^2 + (\delta x_2)^2}$ |
| $|\Delta F| = |F_2 - F_1|$ | $\delta(\Delta F) = \sqrt{(\delta F_1)^2 + (\delta F_2)^2}$ |
| $|\Delta F| = g\,|\Delta m|$ | $\delta(\Delta F) = g \cdot \delta(\Delta m)$ |
| $F = g \cdot m$ | $\delta F = g \cdot \delta m$ |
| $|\Delta m| = |m_2 - m_1|$ | $\delta(\Delta m) = \sqrt{(\delta m_1)^2 + (\delta m_2)^2}$ |
